## Task 2 Solution

```python
# define the filepath for the data to be loaded
# local
datapath = "../../physics725_data/data_week4/ice_data/"
# remote 
#datapath = "/home/share/Project_2/ice_data/"

# create a list of the files 
filelist = np.sort(glob.glob(f"{datapath}/2*"))

# initialize data lists
data = []
times = []
pixels_above_50 = []


for files in filelist:                                            
    # read in the current file
    new_data = np.load(files) 
                                                                     
    # add the data that was just loaded to the end of the "data" array
    data.append(new_data)
                                       
    # get the filename from the file that was just loaded and cut of the .npy ending
    filename = os.path.basename(files)                              
    date = filename.split(".")[0]                                   

    # Extract the digits corresponding to year, month and day from the filename
    year = int(date[:4])                                            
    month = int(date[4:6])                                          
    day = int(date[6:8])  
                                           
    # calculate fraction of the year for each file 
    year_frac = year + (month - 1) / 12 + (day - 1) / 360           
    
    # add the year fraction from the current file to the array times
    times.append(year_frac)                                         
    
    # create a mask to check if entry > 50 
    mask50 = np.logical_and(new_data > 50, ~np.isnan(new_data))     
    # sum up all pixels for which the previous mask has the output True
    count = np.sum(mask50)                                          
    # add the number for each datafile to the array pixels_above_50
    pixels_above_50.append(count)                                 

# convert lists into a numpy arrays
times = np.array(times)                                             
pixels_above_50 = np.array(pixels_above_50)                         

# split the times into data before and after 2012
times_low = times[times<2012]
times_big = times[times>2012]

# split the pixels covered in more than 50% in ice into data before and after 2012
pixels_above_50_low = pixels_above_50[times<2012]
pixels_above_50_big = pixels_above_50[times>2012]

# create a plot
fig, ax = plt.subplots(1,1, figsize=(15,5))

# plot both data sets before and after 2012
ax.plot(times_low, pixels_above_50_low, 'o-', color='tab:blue', ms =4)
ax.plot(times_big, pixels_above_50_big, 'o-', color='tab:blue', ms =4)

# labels and title
ax.set_xlabel('year')
ax.set_ylabel('Pixels covered more than 50% in ice')
ax.set_title('Ice coverage over time')
ax.grid()
```

## Solution to Task 3 

```python
import glob
import os

# Define path to the data
datapath = "'../../physics725_data/data_week4/ice_data/"
filelist = np.sort(glob.glob(f"{datapath}/*.npy"))

# Load the pixel area data (in km^2)
area_per_pixel = np.load("../../physics725_data/data_week4/ice_data_area.npy")
print(np.shape(area_per_pixel))
#~/Documents/Uni/Master/semester2/physics725_data/data_week4/ice_data_area.npy"

# Prepare arrays to store results
times = []
pixels_above_50 = []
area_above_99 = []
total_ice_area = []

# Loop over each data file
for filepath in filelist:
    filename = os.path.basename(filepath)  # e.g., '20030215.npy'
    date_str = filename.split(".")[0]      # remove extension
    year = int(date_str[:4])
    month = int(date_str[4:6])
    day = int(date_str[6:8])
    
    # Convert to fractional year
    fractional_year = year + (month - 1) / 12 + (day - 1) / 360
    times.append(fractional_year)

    # Load ice concentration data
    data = np.load(filepath)
    
    # Count pixels with >50% ice concentration
    valid_mask_50 = data > 50
    pixels_above_50.append(np.sum(valid_mask_50))
    
    # Compute area with >99% ice concentration
    valid_mask_99 = data >= 99
    area99 = np.nansum(area_per_pixel[valid_mask_99])
    area_above_99.append(area99)

    # Compute total ice-covered area (non-nan and >0)
    ice_mask = np.logical_and(~np.isnan(data), data > 0)
    total_area = np.nansum(area_per_pixel[ice_mask])
    total_ice_area.append(total_area)

# Convert results to numpy arrays
times = np.array(times)
pixels_above_50 = np.array(pixels_above_50)
area_above_99 = np.array(area_above_99)
total_ice_area = np.array(total_ice_area)

# --- Plot number of pixels with >50% concentration ---
plt.figure(figsize=(10, 5))
plt.plot(times, pixels_above_50, label="Pixels with >50% ice")
plt.xlabel("Year")
plt.ylabel("Number of Pixels")
plt.title("Pixels with >50% Ice Concentration Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot area with >99% ice concentration ---
plt.figure(figsize=(10, 5))
plt.plot(times, area_above_99, label="Area with >99% ice", color="blue")
plt.xlabel("Year")
plt.ylabel("Area (km²)")
plt.title("Area with >99% Ice Concentration Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Plot total ice-covered area ---
plt.figure(figsize=(10, 5))
plt.plot(times, total_ice_area, label="Total Ice-Covered Area", color="green")
plt.xlabel("Year")
plt.ylabel("Area (km²)")
plt.title("Total Ice-Covered Area Over Time")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
